> \>\>\> c() <br/>

- 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
> \>\>\> c = c + 5 <br/>
> \>\>\> d = d - 5 <br/>
> \>\>\> print(c) <br/>
> [Count (step: 1)] 17 <br/>
> \>\>\> print(d) <br/>
> [Count (step: 2)] 9 <br/>
- 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
> \>\>\> print(c) > 10 <br/>
> True<br/>
> \>\>\> print(d) > 10 <br/>
> False <br/>
> \>\>\> print(c) < 10 <br/>
> False <br/>
> \>\>\> print(d) < 10 <br/>
> True <br/>
> \>\>\> print(c) == 17 <br/>
> True <br/>
> \>\>\> print(d) != 9 <br/>
> False <br/>

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return "[Count (step: %d)] %d" % (self.step, self.count)
    def incr(self):
        self.count = self.count + self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        self.count += other
        return self
    def __sub__(self, other):
        self.count -= other
        return self
    def __lt__(self, other):
        return self.count < other
    def __ne__(self, other):
        return self.count != other
    def __gt__(self, other):
        return self.count > other
    def __eq__(self, other):
        return self.count == other
```

```python
c = Counter(10)
d = Counter(10, 2)
print(c)
print(d)
c.incr()
d.incr()
print(c)
print(d)
c()
d()
print(c)
print(d)
c = c + 5
d = d - 5
print(c)
print(d)
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

3. 직접 호출의 경우는 call 함수에서 합니다. 문제에서는 호출을 하면 step만큼 count를 올려주는데, 이는 incr 함수와 똑같으므로 call 함수 안에 incr()함수를 사용합니다.
4. incr() 함수는 step만큼 count를 올려주는 기능을 합니다. step = step + count로 하면 됩니다.
5. +, - 연산자 오버라이딩 기능을 구현하려면 add, sub 함수를 사용하면 되며, 이는 클래스 내부의 count 값에 other를 대입하도록 해줍니다.

## 문제 4

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

- MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = s

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### 문제 4 해설

1. \_\_init\_\_() 함수는 MySet이란 클래스가 새로 만들어질 때 동작하는 함수로, 여기에서 값을 초기화해준다고 보면 됩니다. 현재 init 함수 내부에서 돌아가는 코드에 대해 좀 더 설명하자면 현재 Myset은 list를 서브클래싱되어있는 상태이므로, MySet은 현재 list 자료형 자체라고 하면 됩니다. 그러므로 self를 가져오면 list 객체를 가져온다고 하면 됩니다. 그리고 init에서 입력되는 값은 내부의 self와 외부에서 가져온 l이란 변수가 있습니다. l도 self와 마찬가지로 list 자료형입니다. Set 객체는 중복을 허용하지 않기에, 모든 값을 모은 다음에 중복된 값을 제거해주면 Set이랑 비슷하다고 할 수 있습니다. init 함수에서는 l 변수에 있는 값을 for문을 이용하여 하나씩 가져와서 self에 하나씩 append합니다. 이렇게 해서 모든 값을 self에 넣었다면 이제 중복값을 제거합니다. 이는 eliminate_duplicate() 함수를 사용해서 했는데, 이 함수에 대해서는 다음 항목에 설명하도록 하겠습니다.

2. eliminate_duplicate()함수는 list에 있는 중복값을 제거해주는 함수로, 여기서 @staticmethod란 데코레이터가 붙는데 말 그대로 정적 메소드라는 걸 컴퓨터에게 알려주는 거라고 보시면 됩니다. 기존 함수와의 차이를 보기 위해서는 입력받는 인자를 잘 보면 됩니다. 일반 함수의 경우는 self 인자가 무조건 있어야 동작합니다만, 이 함수는 self란 인자가 없습니다. 정적 메소드는 클래스에서 직접 접근이 가능하며, 이는 객체와 독립적으로 동작이 가능합니다. 이 함수는 먼저 빈 array를 하나 추가한 다음 입력받은 list를 for문으로 하나씩 가져와서 l에 해당하는 값이 없으면 대입하고 있으면 패스해서 겹치는 값을 제외시킵니다. 이렇게 s에는 중복된 값을 제외한 list가 저장되어있습니다. 그러면 이제 입력된 l에 이 s를 대입하면 중복을 제외한 list가 입력된 l에 대입되게 됩니다.

3. \_\_str\_\_() 함수는 메인 함수에서 해당 클래스 객체를 print()하였을 때 어떻게 표시할 것인지 결정해주는 함수입니다. 이는 list 객체를 하나씩 for문으로 분리하고 이를 양식에 맞게 string으로 배치해주고, 마지막 부분에 불필요한 부분(" ,")을 제거해주고 뒤에 중괄호를 표시해주어서 문제에서 주어진 양식에 맞게 만들어졌다면 해당 값을 return시켜줍니다.

## 문제 5

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
> \>\>\> u = s | t <br/>
> \>\>\> print(u) <br/>
> MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/>

- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
> \>\>\> u = s & t <br/>
> \>\>\> print(u) <br/>
> MySet: {2, 3} <br/>

- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
> \>\>\> s = MySet([1, 2, 3]) <br/>
> \>\>\> t = MySet([3, 4, 5]) <br/>
> \>\>\> u = s - t <br/>
> \>\>\> print(u) <br/>
> MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    # 합집합(|)
    def __or__(self, other):
        self = self + other # 먼저 모든 리스트를 합친 다음
        MySet.eliminate_duplicate(self) #함수를 이용하여 중복값을 제거
        return MySet(self) #클래스 객체 그대로 리턴
    
    # 교집합
    def __and__(self, other):
        l = [] # 먼저 빈 리스트를 만들어본 다음에
        for i in other: # 외부에서 받은 리스트값을 하나씩 가져오면서
            if i in self: # MySet에 해당 값이 있을 경우에
                l.append(i) # 빈 리스트에 값을 추가(self와 other에 있는 값들만 l에 추가됨)
        return MySet(l)
    
    # 차집합
    def __sub__(self, other):
        for i in other: # 외부에서 받은 리스트값을 하나씩 가져오면서
            if i in self: # MySet에 해당 값이 있을 경우에
                self.remove(i) # self에 있는 값을 삭제(self와 other와 겹치는 값들이 제거됨.)
        return MySet(self)
```

```python
s = MySet([1, 2, 2, 3])
print(s)
t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
print(t)
u = s | t
print(u)
u = s & t
print(u)
s = MySet([1, 2, 3]) 
t = MySet([3, 4, 5]) 
u = s - t
print(u)
```

   - 6) getWordsFrequency() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전을 반환하시오.
     - 만약 등록된 URL이 없다면 공백 사전을 반환 
     
     > \>\>\> w1.getWordsFrequency() <br/>
     > {'hello': 8, 'site': 12, 'world': 2, 'science': 11, 'program': 1, 'python': 1}
   - 7) getMaxFreqencyWords() 메소드를 구현하여 각 URL의 웹페이지들을 종합적으로 분석한 단어 출현 빈도 사전에서 가장 많이 출현한 단어 리스트를 반환하시오.
       
     > \>\>\> w1.getMaxFreqencyWords() <br/>
     > site <br/>
     
     - 최다 출현 단어의 빈도수가 동일한 경우 모두 출력해주어야 함 
     
     > \>\>\> w2.getMaxFreqencyWords() <br/>
     > site <br/>
     > science
     
     - 만약 등록된 URL이 없다면 None을 반환
   - 8) 임의의 단어 1개를 파라미터로 받는 searchUrlByWord() 메소드를 구현하여 유사도가 높은 웹 사이트를 출력하시오.
     > \>\>\> w1.searchUrlByWord("news")<br/>
     > http://www.cnn.com <br/>
     
     - 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.  

```python
import string
from urllib import request
class WebWordsFrequency:
    
    
    def __init__(self, *urls):
        self.dic = {}
        for url in urls:
            self.findWords(url)
        
    def addUrl(self, url):
        self.findWords(url)
        
    def findWords(self, url):
        stopword = []
        with open('stopword.txt', 'r', encoding='utf8') as f:
            for line in f:
                if (line != "\n"):
                    stopword.append(line.strip())
        stopword.append("nbsp")
        stopword.append("lt")
        stopword.append("gt")
        stopword.append("amp")
        stopword.append("quot")
        stopword.append("pm")
        stopword.append("et")
        stopword.append("est")
        stopword.append("raquo")
        
        total_word = 0
        words = []
        javascript_check = False
        css_check = False

        source_bytes = request.urlopen(url).read()
        source = source_bytes.decode('utf-8')

        for i in source.split("<"):
            m = i.split(">")

            if len(m) > 1:
                check = m[0].split()
                if len(check) > 0:
                            # if('script' in m[0] and m[0].find("description") == -1 and m[0].find("subscription") == -1 and m[0].find("transcript") == -1 and m[0].find(" javascript:void();") == -1): javascript_check = True
                    if ('script' == check[0]): javascript_check = True
                    if ('/script' in m[0]): javascript_check = False
                            # if(m[0].find('style') != -1 and m[0].find('style=') == -1 and m[0].find('stylesheet') == -1and m[0].find('Lifestyle') == -1): css_check = True
                    if ('style' == check[0]): css_check = True
                    if ('/style' in m[0]): css_check = False
                    m = m[1].strip().lower()
                    if (m != "" and not javascript_check and not css_check):
                        rep = ["//", "--", '\'', '$', '©', '&', '%', '"', '^', 'ⓒ', '·', '…', '—', '~', '–', '»', '!',
                                       '”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_', '“', '‘', '’', '.', ',']
                        rep = rep + list(string.punctuation)
                        rep = list(set(rep))
                        for r in rep:
                            m = m.replace(r, ' ')
                        m = m.split()
                                # print(m)
                        for n in m:
                            if (n != ''):
                                if (n.isdigit()):
                                    continue
                                else:
                                    if (n not in stopword):
                                        total_word += 1
                                        words.append(n.strip())
        self.dic[url] = words
        
    def __str__(self):
        return str(self.dic)
    
    def removeUrl(self, url):
        del self.dic[url]
        
    def listUrls(self):
        print("\n".join(self.dic.keys()))
    
    def getWordsFrequency(self):
        t = {}
        l = []
        for i in list(self.dic.values()):
            l = l + i
        for i in l:
            if (i not in t.keys()):
                t[i] = 1
            else:
                t[i] = t[i] + 1
        print(t)
        return t

    def getMaxFreqencyWords(self):
        t = self.getWordsFrequency()
        if(t != {}):
            high = max(t.items(), key= lambda d: d[1])[1]
            l = [k for k, v in t.items() if v == high]
            print("\n".join(l))
            
    def searchUrlByWord(self, s):
        l = []
        for k,v in self.dic.items():
            l.append((k,v.count(s)))
        print(max(l, key= lambda d: d[1])[0])
```

```python
w1 = WebWordsFrequency('http://www.github.com', 'http://www.times.com')
w1.addUrl('http://www.python.org/')
w1.removeUrl('http://www.github.com')
w1.listUrls()
w1.getWordsFrequency() 
w1.getMaxFreqencyWords()
w1.searchUrlByWord("python")
```

## 문제 8

위 7번 문제에서 정의한 WebWordsFrequency 클래스를 상속하여 OrderedWebWordsFrequency 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 오버라이드 하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
   - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
   - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
     - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
     - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
     
   > \>\>\> w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
     
   > \>\>\> w4.getWordsFrequency()<br/>
   > [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
     
   > \>\>\> w4.getWordsFrequency(reverse=true)<br/>
   > [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse = False):
        l = []
        t = {}
        for i in list(self.dic.values()):
            l = l + i
        for i in l:
            if (i not in t.keys()):
                t[i] = 1
            else:
                t[i] = t[i] + 1
        l = list(t.items())
        l.sort(key=lambda d:d[1], reverse = not reverse)
        self.o_freq = l
        t = dict(l)
        self.freq = t
        print(self.o_freq)
```

```python
w4 = OrderedWebWordsFrequency('http://www.times.com', 'https://github.com')
w4.getWordsFrequency()
w4.getWordsFrequency(reverse=True)
```

#### 문제8 해설

OrderedWebWordsFrequency 클래스는 기존 WebWordsFrequency 클래스에 정렬 기능이 추가된 클래스입니다. 먼저 기존 WebWordsFrequency 클래스에서 했던 기능을 그대로 가져옵니다. 차이점은 정렬을 위해 사전 자료형을 list로 바꿔주는데 이는 t.items()로 키와 value 모두 가져옵니다. 이렇게 가져오게 되면 [(Word, Frequency), (Word, Frequency), ...] 형태로 받아집니다. 그러면 이제 이 객체를 value값을 기준으로 정렬을 시켜줍니다. 이는 sort(key=lambda d:d[1])로 해줍니다. 그리고 추가로 reverse를 입력받는데, 기본 값은 False로 해주고, sort에서 sort(key=lambda d:d[1], reverse = not reverse)로 해주면 reverse도 입력받은 대로 할 수 있습니다. 그리고 정렬한 결과를 그대로 출력해주면 이 함수의 기능은 모두 구현했다고 할 수 있습니다.

## 문제 9

다음과 같은 코딩이 가능하도록 OrderedWebWordsFrequency 안에 반복자와 관련된 메소드를 추가하시오.

```python
class OrderedWebWordsFrequency(WebWordsFrequency):
    def getWordsFrequency(self, reverse = False):
        l = []
        t = {}
        for i in list(self.dic.values()):
            l = l + i
        for i in l:
            if (i not in t.keys()):
                t[i] = 1
            else:
                t[i] = t[i] + 1
        l = list(t.items())
        l.sort(key=lambda d:d[1], reverse = not reverse)
        self.o_freq = l
        t = dict(l)
        self.freq = t
        print(self.o_freq)
        

    def __iter__(self):
        return iter(self.o_freq)
        
```
