#### 해설 : 문제 2

문자열이 입력되면 먼저 이 문자열을 띄어쓰기(' ')를 기준으로 분리해서 단어의 뭉치로 만들어줍니다.

그 다음 이 순서를 역으로 하기 위해 reverse 함수를 이용합니다.

````python
r.reverse()
print(r)
````

역순으로 배열이 되었다면 이제 list를 하나의 string으로 만들어주는데, 이때는 join을 사용합니다. join 앞에 들어가는 건 list안의 객체가 하나로 합쳐실 때 그 사이에 어떤 것을 넣게 할 지 정해주는 것입니다. 여기서는 띄어쓰기(' ')를 넣어서 순서가 역순으로 된 문자열이지만 단어 하나하나는 그대로인 상태로 만듭니다.

````python
r = " ".join(r)
print(r)
````

그 다음 리스트에서 맨 마지막 요소만 따로 가져오고, 맨 마지막 요소를 제외한 리스트는 다시 원래대로 돌려야하는데, 이 때는 join을 이용한다. 이렇게 하면 리스트가 문자열로 다시 뭉쳐지며, 뭉쳐지는 과정에서 리스트 요소 사이에는 "/"이 삽입된다. 그리고 새로운 list 변수 l을 만든 다음에 append를 사용하여 디렉토리, 파일명 하나씩 넣는다. 그리고 list를 출력하면 원하는 결과가 출력된다.

````python
t = r.split(" ")
t = "".join(t)
print(t)
````

다음 문자열을 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오 (즉, 문자열 S에서 l을 만들어라)

```
s = '  first star   :   second star   :    third star  '
l = ['first star', 'second star', 'third star']
```

2) 리스트 내포(list comprehension)를 이용한 코드를 작성하시오.

## 문제 5

다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
```
>>> addall([1])
1
>>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
45
```

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
```
>>> addallodd([1])
1
>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
25
```

- [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
- [참고 2]: sum() 등의 내장함수 사용 가능함

````python
def addall(lst):
    return sum([i for i in lst])

print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addallodd(lst):
    return sum([i for i in lst if i % 2 == 1])

print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
a = "26"
if (len(a) == 1): a = "0" + a
a = list(a)
m = [0,0]

i = 0
for t in a:
    a[i] = int(a[i])
    m[i] = int(a[i])
    i = i + 1
print("a = " + str(a))
print("m = " + str(m))
````

이제 while문을 이용하는데, 그냥 a != m으로 판단하면, 아직 while문에 들어가기 전까지는 a랑 m이랑 같기에 loop에 들어가지 못한다. 그래서 이를 위해 or i == 0으로 하여 처음일 때는 무시하게 한다. 그 다음엔 문제에 설명한 것과 같이 1의 자리의 숫자는 10의 자리로 옮기고, 1의 자리는 바뀌기 전 1의 자리와 10의 자리를 더한 값을 넣는데, 10을 초과하면 10을 빼서 해당 값의 1의 자리만 가져오게 한다. 이를 계속 반복하면 몇 번 사이클을 돌렸는지는 i 변수에 저장되므로 출력할 때는 i를 출력하면 된다.

````python
i = 0

while(a != m or i == 0):
    tmp = a[0]
    a[0] = a[1]
    a[1] = a[0] + tmp
    if(a[1] > 9): a[1] -= 10
    print("==============")
    print("a = " + str(a))
    print("m = " + str(m))
    i = i + 1
    
print("사이클 횟수 = " + str(i))
````

## Incremental Project

[Incremental Project] 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.
- 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
    - 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
    - 모든 punctuation을 확인하는 방법은 아래와 같다.
``` python
    import string
    print(string.punctuation)
```
- 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
    - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']
- 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
    - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}
- 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
    - http://www.naver.com
    - http://www.daum.net
    - http://www.nytimes.com

````python
import string
from urllib import request 

urls = ['http://www.naver.com', 'http://www.daum.net', 'http://www.nytimes.com']
for url in urls:
    total_word = 0
    words = []
    dic = {}
    javascript_check = False
    css_check = False

    f = request.urlopen(url)
    source_bytes = f.read()
    source = source_bytes.decode('utf-8')

    for i in source.split("<"):
        m = i.split(">")
        
        if('script' in m[0]): javascript_check = True    
        if('/script' in m[0]): javascript_check = False  
        if('style type="text/css"' in m[0]): css_check = True   
        if('/style' in m[0]): css_check = False
                
        if len(m) > 1:
            m = m[1].strip()
            if(m != "" and not javascript_check and not css_check):
                rep = ["//","--",'\'', '$','©','&', '%', '"', '^', 'ⓒ', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
                rep = rep + list(string.punctuation)
                rep = list(set(rep))
                for r in rep:
                    m = m.replace(r, ' ')
                m = m.split()
                for n in m:
                    if(n != ''): 
                        if(n.isdigit()): continue
                        else:
                            total_word += 1
                            words.append(n.strip())
    for i in words:
        if(i not in dic):
            dic[i] = 1
        else:
            dic[i] = dic[i] + 1
   
    print(dic)
    print()
````

#### 해설 : Incremental Project

먼저 source에서 "<"로 split을 하게 되면 다음과 같이 나온다.

````python
print(source.split("<"))
````

그리고 이 상태에서 for문을 이용해 하나씩 분리시킨 다음 이를 또 ">"로 split을 하게 되면 다음과 같이 나온다.

````python
for i in source.split("<"):
    m = i.split(">")
    print(m)
````

이렇게 되면 배열마다 m[0]은 tag 안의 내용으로, m[1]은 태그 밖의 내용이 나온다.
그런데 그냥 m[1]을 가져오게 되면, 맨 첫번째 배열의 크기가 1이기에 오류가 난다. 그래서 이를 방지하기 위해 크기가 2부터인 것만 받는다.
그리고 m[1]을 출력하다보면 \n, \t 등 공백 문자들이 있기에 이는 strip()으로 공백을 제거한 다음, m[1]이 비어있지 않은 것만 받는다.

이렇게 하면 일단 HTML 태그를 제외한 순수 텍스트 문자열은 받게 된다. 그러나, 이 과제에서는 순수 텍스트 문자에서 단어의 갯수를 알아내야 하는데, 여기에서 특수문자를 제외해야한다. 그렇기에 특수문자를 하나씩 제거하는데, 단어 구분을 위해 특수문자들은 띄어쓰기(" ")로 바꾸기로 하였다. 그 다음에 띄어쓰기(" ")로 split을 하면 이제 한 단어마다 list에 저장이 된다. 그럼 이제 비어있는 값과 숫자를 빼면 한 단어씩 출력이 되는 것이다.

````python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = list(string.punctuation)
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                n = n.strip()
                if(n != ''): 
                    if(n.isdigit()): continue
                    else:
                        total_word += 1
                        print(n)
````

그러면 남은 건 한 단어씩 가져왔으면 total_word에 1씩 더해주면 총 단어 수가 된다.

````python
print(total_word)
````
