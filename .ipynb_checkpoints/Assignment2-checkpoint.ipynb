```python
from urllib import request 
 
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
```

```python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = ["//","--", '$','©','&', '%', '"', '^', '·', '~', '!', '㈜', '-', '/', ':', '(', ')', '[', ']', '_', '‘', '’', '.', ',']
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                if(n != ''): 
                    total_word += 1
                    print(n)
print("총 단어 수 = " + str(total_word) + "개")
```
