## 문제 2

경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - div_path('/usr/local/bin/python')
        - ['/usr/local/bin', 'python']
    - div_path('/home/chulsoo/test.txt')
        - ['/home/chulsoo', 'test.txt']
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.

````python
def div_path(s):
    t = s.split("/")
    l = []
    l.append("/".join(t[:-1]))
    l.append(t[-1])
    return l

s = input()
print(div_path(s))
````

#### 해설 : 문제 2

## 문제 3
두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.
- 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
- 반환하는 리스트는 정렬이 되어 있어야 한다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - list_union([1, 2, 3], [1, 2, 4])
        - [1, 2, 3, 4]
    - list_union([-10, -5, 0, -1], [100, 9, 0, 9])
        - [-10, -5, -1, 0, 9, 100]
    - list_union([0, 1, 2], [0, 1, 2])
        - [0, 1, 2]
- [참고] 리스트(l) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 l.append(10) 이다.
- [참고] 임의의 정수값 (x)이 리스트 (l) 내에 존재하는지 판단하는 방법은 x in l 이다.

````python
def list_union(lista, listb):
    lst = []
    
    for i in lista:
        lst.append(i)
        
    for i in listb:
        if (i not in lst): lst.append(i)
    lst.sort()        
    return lst

print(list_union([-10, -5, 0, -1],  [100, 9, 0, 9]))
````

````python
lst = []
lista = [-10, -5, 0, -1]
listb = [100, 9, 0, 9, -1]
for i in lista:
    lst.append(i)
        
for i in listb:
    if (i not in lst): lst.append(i)
print(lst)
````

````python
lst.sort()
print(lst)
````

26부터 시작한다. 2+6 = 8이다. 새로운 숫자는 68이다. 6+8 = 14이다. 새로운 숫자는 84이다. 8+4 = 12이다. 새로운 숫자는 42이다. 4+2 = 6이다. 새로운 숫자는 26이다.

N이 주어졌을 때, N의 사이클의 길이를 구하는 프로그램을 작성하시오.

````python
a = input()
if (len(a) == 1): a = "0" + a
a = list(a)
m = [0,0]

i = 0
for t in a:
    a[i] = int(a[i])
    m[i] = int(a[i])
    i = i + 1
i = 0

while(a != m or i == 0):
    tmp = a[0]
    a[0] = a[1]
    a[1] = a[0] + tmp
    if(a[1] > 9): a[1] -= 10
    i = i + 1

print(i)
````

## Incremental Project

웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그를 제외한 순수 텍스트 문자열만을 걸러내고, 2) 그 순수 텍스트 문자열 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.

- HTML 태그는 다음 조건을 가지는 것이다.

    - < 로 시작하여 >로 끝난다.
    - 예를 들어, 아래와 같은 HTML 내용 내에 한 줄이 있다고 가정하자.
    ```html
    <label for="keep_signed">로그인 유지</label>
    ```
    - 위 소스에서 HTML 태그를 제외한 순수 텍스트 문자열은 아래와 같다.
    로그인 유지
    - 한편, 아래와 같은 HTML 한 줄이 있다고 가정하자.
    ```html
    <link rel="stylesheet" href="/common/css/xe.min.css?20150910092654" />
    ```
    - 위 소스는 HTML 태그로만 구성된 것이며 걸러낼 수 있는 순수 텍스 문자열은 존재하지 않는다.
- 단어를 나누는 기준은 오로지 공백문자 (whitespace)이다.

    - 즉, 위 순수 텍스트 문자열인 "로그인 유지"에서 분리한 단어는 "로그인"과 "유지"이다.
- 중복을 고려하지 않고 무조건 모든 단어의 개수를 세어서 출력한다.
- 본 숙제를 위하여 HTML 파싱과 관련된 모듈 (예. BeautifulSoup) 사용 금지
- 다음은 URL로 지정된 웹페이지를 특정 문자열로 가져오는 코드이다.
    - 즉, 숙제 코드는 아래 코드로 부터 시작해야 한다.

````python
import string
from urllib import request 
 
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
print(list(string.punctuation))
````

````python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = ["//","--",'\'', '$','©','&', '%', '"', '^', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                if(n != ''): 
                    if(n.isdigit()): continue
                    else:
                        total_word += 1
print("총 단어 수 = " + str(total_word) + "개")
````

#### 해설 : Incremental Project

먼저 source에서 "<"로 split을 하게 되면 다음과 같이 나온다.

````python
print(source.split("<"))
````

그리고 이 상태에서 for문을 이용해 하나씩 분리시킨 다음 이를 또 ">"로 split을 하게 되면 다음과 같이 나온다.

````python
for i in source.split("<"):
    m = i.split(">")
    print(m)
````

이렇게 되면 배열마다 m[0]은 tag 안의 내용으로, m[1]은 태그 밖의 내용이 나온다.
그런데 그냥 m[1]을 가져오게 되면, 맨 첫번째 배열의 크기가 1이기에 오류가 난다. 그래서 이를 방지하기 위해 크기가 2부터인 것만 받는다.
그리고 m[1]을 출력하다보면 \n, \t 등 공백 문자들이 있기에 이는 strip()으로 공백을 제거한 다음, m[1]이 비어있지 않은 것만 받는다.

이렇게 하면 일단 HTML 태그를 제외한 순수 텍스트 문자열은 받게 된다. 그러나, 이 과제에서는 순수 텍스트 문자에서 단어의 갯수를 알아내야 하는데, 여기에서 특수문자를 제외해야한다. 그렇기에 특수문자를 하나씩 제거하는데, 단어 구분을 위해 특수문자들은 띄어쓰기(" ")로 바꾸기로 하였다. 그 다음에 띄어쓰기(" ")로 split을 하면 이제 한 단어마다 list에 저장이 된다. 그럼 이제 비어있는 값과 숫자를 빼면 한 단어씩 출력이 되는 것이다.

````python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = ["//","--",'\'', '$','©','&', '%', '"', '^', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                if(n != ''): 
                    if(n.isdigit()): continue
                    else:
                        total_word += 1
                        print(n)
````

그러면 남은 건 한 단어씩 가져왔으면 total_word에 1씩 더해주면 총 단어 수가 된다.

````python
print(total_word)
````

ACM-ICPC 문제의 경우는, 실수했던 부분은 리스트 안의 값을 string에서 int로 변경하지 못해서 더했을 때 그냥 뒤에 값이 붙여져서 list 안의 값을 모두 int로 변경해서 해결하였습니다. 그래도 속도면에서 나쁘지 않았던 것 같아서 다행입니다. 나중에 코드를 더 압축하거나 더 효율적인 방법을 생각해봐야할 것 같습니다.

Incremental Project는 시간이 오래 걸리긴 했지만, 그만큼 보람찼던 문제였습니다. 기존에 크롤링을 했을 때 주로 ruby에서 nokogiri를 사용했고, 전체 소스를 String으로 해서 언어 내부 String 기능으로 원하는 정보를 가져오는 건 처음인 것 같습니다. 사실 split으로 꺽쇠 기호가 원하는 대로 분리될까 걱정했는데 예상외로 원하는 대로 분리가 되어서 다행이었습니다. 그래도 실제 사이트 주소 코드를 가져와서 변수가 많았습니다. 특히 특수문자에서 애를 많이 먹었습니다. 특수문자를 인식하는 함수가 있는지 찾아봤는데 외부 함수인 re를 사용하는 방법 등이 있었지만 결국 특수 문자를 보관하는 리스트를 하나 만들어서 for문으로 특수문자를 하나씩 replace해서 제거하였습니다. 그리고 기존에 완성한 코드는 숫자까지 단어로 포함했었는데, 혹시나 몰라서 숫자는 제외시켰습니다.

과제하면서 쥬피터 노트북과 마크다운 사용이 조금 적응이 된 것 같습니다. 확실히 쥬피터 노트북이 포트폴리오화하기 좋은 툴인 것 같습니다. 그래서 쥬피터 노트북을 활용하는 방법도 찾아봐야할 것 같습니다.
