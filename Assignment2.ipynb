## ACM-ICPC 문제 1110번

0보다 크거나 같고, 99보다 작거나 같은 정수가 주어질 때 다음과 같은 연산을 할 수 있다. 먼저 주어진 수가 10보다 작다면 앞에 0을 붙여 두 자리 수로 만들고, 각 자리의 숫자를 더한다. 그 다음, 주어진 수의 가장 오른쪽 자리 숫자와 앞에서 구한 합의 가장 오른쪽 자리 숫자를 이어 붙이면 새로운 수를 만들 수 있다. 다음 예를 보자.

26부터 시작한다. 2+6 = 8이다. 새로운 숫자는 68이다. 6+8 = 14이다. 새로운 숫자는 84이다. 8+4 = 12이다. 새로운 숫자는 42이다. 4+2 = 6이다. 새로운 숫자는 26이다.

위의 예는 4번만에 원래 숫자로 돌아올 수 있다. 따라서 26의 사이클의 길이는 4이다.

N이 주어졌을 때, N의 사이클의 길이를 구하는 프로그램을 작성하시오.

```python
a = input()
if (len(a) == 1): a = "0" + a
a = list(a)
m = [0,0]

i = 0
for t in a:
    a[i] = int(a[i])
    m[i] = int(a[i])
    i = i + 1
i = 0

while(a != m or i == 0):
    tmp = a[0]
    a[0] = a[1]
    a[1] = a[0] + tmp
    if(a[1] > 9): a[1] -= 10
    i = i + 1

print(i)
```

### 해설: ACM-ICPC 문제 1110번

```python
a = "26"
if (len(a) == 1): a = "0" + a
a = list(a)
m = [0,0]

i = 0
for t in a:
    a[i] = int(a[i])
    m[i] = int(a[i])
    i = i + 1
print("a = " + str(a))
print("m = " + str(m))
```

먼저 값이 string 형태로 입력 받으면 1의 자리만 있으면 앞에 "0"을 붙여서 10의 자리까지 만들어주고, string을 list로 바꿔서 변환하기 편하게 해준다. 그리고 새로운 리스트 변수 m을 만들어서 나중에 값을 비교할 때 사용한다. 그리고 리스트에 있는 값을 int로 변경해주고, 이 때 m에도 값을 넣어준다. 참고로 m = a로 하면 안 되는 이유는, 이렇게 하면 나중에 a의 값이 변경될 때 같이 변경되기 때문이다.

```python
i = 0

while(a != m or i == 0):
    tmp = a[0]
    a[0] = a[1]
    a[1] = a[0] + tmp
    if(a[1] > 9): a[1] -= 10
    print("==============")
    print("a = " + str(a))
    print("m = " + str(m))
    i = i + 1
    
print("사이클 횟수 = " + str(i))
```

이제 while문을 이용하는데, 그냥 a != m으로 판단하면, 아직 while문에 들어가기 전까지는 a랑 m이랑 같기에 loop에 들어가지 못한다. 그래서 이를 위해 or i == 0으로 하여 처음일 때는 무시하게 한다. 그 다음엔 문제에 설명한 것과 같이 1의 자리의 숫자는 10의 자리로 옮기고, 1의 자리는 바뀌기 전 1의 자리와 10의 자리를 더한 값을 넣는데, 10을 초과하면 10을 빼서 해당 값의 1의 자리만 가져오게 한다. 이를 계속 반복하면 몇 번 사이클을 돌렸는지는 i 변수에 저장되므로 출력할 때는 i를 출력하면 된다.

## Incremental Project

```python
from urllib import request 
 
url = 'http://cse.koreatech.ac.kr'
f = request.urlopen(url)
source_bytes = f.read()
source = source_bytes.decode('utf-8')
```

```python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = ["//","--", '$','©','&', '%', '"', '^', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                if(n != ''): 
                    total_word += 1
                    #print(n)
print("총 단어 수 = " + str(total_word) + "개")
```

### 해설 : Incremental Project

먼저 source에서 "<"로 split을 하게 되면 다음과 같이 나온다.

```python
print(source.split("<"))
```

그리고 이 상태에서 for문을 이용해 하나씩 분리시킨 다음 이를 또 ">"로 split을 하게 되면 다음과 같이 나온다.

```python
for i in source.split("<"):
    m = i.split(">")
    print(m)
```

이렇게 되면 배열마다 m[0]은 tag 안의 내용으로, m[1]은 태그 밖의 내용이 나온다.
그런데 그냥 m[1]을 가져오게 되면, 맨 첫번째 배열의 크기가 1이기에 오류가 난다. 그래서 이를 방지하기 위해 크기가 2부터인 것만 받는다.
그리고 m[1]을 출력하다보면 \m, \t 등 공백 문자들이 있기에 이는 strip()으로 공백을 제거한 다음, m[1]이 비어있지 않은 것만 받는다.

이렇게 하면 일단 HTML 태그를 제외한 순수 텍스트 문자열은 받게 된다. 그러나, 이 과제에서는 순수 텍스트 문자에서 단어의 갯수를 알아내야 하는데, 여기에서 특수문자를 제외해야한다. 그렇기에 특수문자를 하나씩 제거하는데, 단어 구분을 위해 특수문자들은 띄어쓰기(" ")로 바꾸기로 하였다. 그 다음에 띄어쓰기(" ")로 split을 하면 이제 한 단어마다 list에 저장이 된다. 그럼 이제 비어있는 값만 빼면 한 단어씩 출력이 되는 것이다.

```python
total_word = 0
for i in source.split("<"):
    m = i.split(">")
    if len(m) > 1:
        m = m[1].strip()
        if(m != ""):
            rep = ["//","--", '$','©','&', '%', '"', '^', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split(" ")
            for n in m:
                if(n != ''): 
                    total_word += 1
                    print(n)
```

그러면 남은 건 한 단어씩 가져왔으면 total_word에 1씩 더해주면 총 단어 수가 된다.
