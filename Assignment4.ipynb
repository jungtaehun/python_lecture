## 문제 5

다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
```
>>> addall([1])
1
>>> addall([1, 2, 3, 4, 5, 6, 7, 8, 9])
45
```

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)
```
>>> addallodd([1])
1
>>> addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9])
25
```

- [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
- [참고 2]: sum() 등의 내장함수 사용 가능함

````python
def addall(lst):
    sum = 0
    for i in lst:
        sum += i
    return sum

print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addall(lst):
    return sum([i for i in lst])

print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addallodd(lst):
    sum = 0
    for i in lst:
        if i % 2 == 1:
            sum += i
    return sum

print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

````python
def addallodd(lst):
    return sum([i for i in lst if i % 2 == 1])

print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
````

#### 문제5 해설

먼저 for ~ in 방식을 사용할 때에는 sum이란 리터럴을 생성해서 0으로 초기화시킵니다. 그 다음 for ~ in을 통해 값을 하나씩 가져와서 각 값들을 sum에 더해줍니다. 여기서 만약 addalodd 함수의 경우는 홀수만 더해주기로 했기에 for ~ in 안에 if i % 2 == 1로 필터링해서 나온 결과만 sum에 더해줍니다. 그리고 모두 더했다면 sum을 return시킵니다.

다만 이 문제는 리스트 내포 방식으로도 가능합니다. 여기서는 sum()내장함수를 이용하는데, list를 넣어주면 list 내부의 값을 모두 더한 값을 반환해줍니다. 먼저 리스트 내포 방식으로 리스트를 만들면 다음과 같습니다.
```python
[i for i in lst]
```
여기에 홀수만 필터링하고자하면 for ~ in 다음에 if i % 2 == 1을 하나 추가합니다.
```python
[i for i in lst if i % 2 == 1]
```
이렇게 해당하는 list를 만들었다면 이를 sum()내장 함수에 넣어서 이를 return시키면 for ~ in을 사용한 방식보다 더 짧게 구현할 수 있습니다.

- [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.

````python
def sort_val(d):
    return d[1]
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
keys = list(d.keys())
items = list(d.items())
````

#### 문제7 해설

튜플을 키로 sort하는 것은 값을 sort하는 것보다는 간단합니다. 먼저 d.keys()를 통해 key값을 받아서 list에 보관한 다음에, 이를 sort한 다음, for ~ in을 이용해서 sort한 키에 대해서 값을 출력시키면 키가 sort된 상태에서 출력이 됩니다. 여기서 역으로 sort를 하려면 sort함수에 reverse=True로 주면 됩니다.

이제 값을 sort하는데, 대부분 값을 sort하려면 sorted란 외부 함수를 사용하는 방법도 있지만, sort를 사용하면서 이를 구현하려면 함수를 이용합니다. 이번엔 한번 lambda를 이용해보았습니다. 여기서 lambda는 C언어에서 inline함수와 비슷하다고 할 수 있습니다. 예시는 다음과 같습니다.

````python
def add_func(x, y):
    return x+y
print(add_func(2,3))
print((lambda x, y : x+y)(2,3))
````

lambda를 이용하면 간단한 함수를 더 간단하게 구현할 수 있습니다. 그리고 이 lambda를 key값에 넣게 되면 데이터 뭉치에서 원하는 값만 가져와서 그 값에 따른 sort가 가능합니다. 먼저 d.items()을 통해 키와 값이 모두 모여있는 list로 만든 다음, 함수를 이용하여 여기에서 값을 기준으로 이를 sort해줍니다. 역으로 하려면 아까처럼 reverse=True시키면 됩니다. 그리고 이를 for ~ in을 이용해서 출력하면 키에 따라 정렬된 상태로 출력하게 됩니다. 참고로 이 방법은 key로 정렬하는 것도 가능합니다.

## Incremental Project

[Incremental Project] 이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오.
- 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.
    - 예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']
    - 모든 punctuation을 확인하는 방법은 아래와 같다.
``` python
    import string
    print(string.punctuation)
```
- 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.
    - 예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']
- 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
    - 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}
- 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.
    - http://www.naver.com
    - http://www.daum.net
    - http://www.nytimes.com

````python
import string
from urllib import request 

urls = ['http://edition.cnn.com/','https://www.nytimes.com/', 'http://abcnews.go.com/', 'http://www.reuters.com/']
stopword = []
with open('stopword.txt', 'r') as f:
    for line in f:
        if(line != "\n") :
            stopword.append(line.strip())
                
rep = list(set(rep))
                
for url in urls:
    total_word = 0
    words = []
    dic = {}
    javascript_check = False
    css_check = False

    f = request.urlopen(url)
    source_bytes = f.read()
    source = source_bytes.decode('utf-8')

    for i in source.split("<"):
        m = i.split(">")
        
        if('script' in m[0]): javascript_check = True    
        if('/script' in m[0]): javascript_check = False  
        if(m[0].find('style') != -1 and m[0].find('style=') == -1): css_check = True   
        if('/style' in m[0]): css_check = False
                
        if len(m) > 1:
            m = m[1].strip().lower()
            if(m != "" and not javascript_check and not css_check):
                rep = ["//","--",'\'', '$','©','&', '%', '"', '^', 'ⓒ', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
                rep = rep + list(string.punctuation)      
                for r in rep:
                    m = m.replace(r, ' ')
                m = m.split()
                for n in m:
                    if(n != ''): 
                        if(n.isdigit()): continue
                        else:
                            if(n not in stopword):
                                total_word += 1
                                words.append(n.strip())
    for i in words:
        if(i not in dic):
            dic[i] = 1
        else:
            dic[i] = dic[i] + 1
   
    items = list(dic.items())
    items.sort(key=lambda d: d[1], reverse=True)
    for k,v in items:
        print(k,v)
    print()
````

CSS도 마찬가지로 'style type="text/css"'가 들어가면 css_check를 true로, '/style'이 나오면 css_check를 false로 해주면서 아까와 같이 필터링을 해줍니다.

이렇게 하면 일단 HTML 태그, javascript, css를 제외한 순수 텍스트 문자열은 받게 됩니다. 그러나, 이 과제에서는 순수 텍스트 문자에서 단어의 갯수를 알아내야 하는데, 여기에서 특수문자를 제외해야합니다. 그래서 과제 2에서 만든 특수문자 모음에 String에 포함되어있는 punctuation단어를 모두 포함해서 중복되는 부분은 지워줍니다. 이는 set을 이용해서 중복을 삭제하고 다시 list로 변경하였습니다. 그 다음 특수문자를 하나씩 제거하는데, 단어 구분을 위해 특수문자들은 띄어쓰기(" ")로 바꾸기로 하였습니다. 그 다음에 띄어쓰기(" ")로 split을 하면 이제 한 단어마다 list에 저장이 됩니다. 그럼 이제 비어있는 값과 숫자를 빼면 한 단어씩 출력이 됩니다.

````python
words = []
dic = {}
for i in source.split("<"):
    m = i.split(">")
        
    if('script' in m[0]): javascript_check = True    
    if('/script' in m[0]): javascript_check = False  
    if('style type="text/css"' in m[0]): css_check = True   
    if('/style' in m[0]): css_check = False
                
    if len(m) > 1:
        m = m[1].strip()
        if(m != "" and not javascript_check and not css_check):
            rep = ["//","--",'\'', '$','©','&', '%', '"', '^', 'ⓒ', '·','…', '~', '!','”', '㈜', '-', '/', ':', '(', ')', '[', ']', '_','“', '‘', '’', '.', ',']
            rep = rep + list(string.punctuation)
            rep = list(set(rep))
            for r in rep:
                m = m.replace(r, ' ')
            m = m.split()
            for n in m:
                if(n != ''): 
                    if(n.isdigit()): continue
                    else:
                        total_word += 1
                        words.append(n.strip())
                        print(n.strip())
````

이제 이 list를 dictionary로 변환하기 위해, words에 저장되어 있는 단어를 하나씩 가져와서 dic에 없는 단어면 key는 단어로 하는 객체에 값은 1로 추가해주고 dic에 있는 단어이면 1씩 더해줍니다. 이렇게 하면 다음과 같이 dictionary 형태로 변환됩니다.

## 느낀 점

이번 과제는 리스트, 문자열, 사전 등을 중심으로 다룬 과제였습니다. 점점 파이썬이 내부에서 어떻게 돌아가는지 보면서 흥미로운 것 같습니다. 예전엔 단순히 파이썬은 쉬운 언어다 싶었는데, 메모리적인 관점에서 바라봤을 때 독특하게 운영되는 것이 신기합니다. 그리고 이번에 사전의 값에 따른 sort를 찾다가 lambda란 문법을 발견했는데, 확실히 기존 함수를 더 간략하게 표현할 수 있다는 것이 좋은 것 같습니다. 그리고 sort 함수 내에 key를 설정해서 sort가 가능한 것을 보면서 sort란 함수안에는 for문을 이용해서 작동하는 것 같았습니다. 그 때 교수님이 얘기해주셨던 얘기중에 "사전은 sort를 하지 않는 게 좋다"라고 하셨는데, 그 이유를 생각해보면 사전 객체의 목적은 어떤 것을 정렬하는 목적이 아니라 어떤 키를 주면 해당하는 값을 빨리 찾아주는 것이 목적이기 때문이라고 생각했습니다. 굳이 정렬을 한다면 사전보다는 리스트가 더 적합한 객체라서 그런 것 같습니다. 그리고 네이버를 크롤링하면서 느낀 점은 네이버는 대부분 정보를 가져오는 데 javascript로 정보를 가져온다는 것을 알 수 있었습니다. 그래서 예상 외로 네이버에서 javascript를 제외하면 네이버에 들어갈 때 보여주는 단어의 갯수에 비해서 많은 단어를 갖고 있지 않은 것 같습니다. 이것이 웹에서 효율적인 구조인 것이, html에서 반복적인 부분은 javascript나 python같은 스크립트 언어로 for문을 사용해서 일괄적으로 구조를 만들어주는 것이 불필요한 코드를 줄여줄 수 있어서 네이버에서도 이러한 방식을 사용하는 것 같습니다. 그리고 javascript와 css를 제외하는 방법으로 특정 문장이 포함되어있을 때 다른 특정 부분이 나오기 전까지 그 부분을 제외하는 걸로 구현하였는데, 이는 너무 비효율적이고 모든 상황에 들어맞지는 않아서 더 효율적인 방법이 무엇이 있는지 잘 모르겠습니다. 그래도 파이썬을 하면서 예전에 배웠던 ruby와 비교하면 서로의 장단점이 있는 것 같고, 파이썬의 내부 구조를 보면서 ruby의 내부 구조는 파이썬이랑 다른지도 궁금할 때도 있습니다. 또한 ruby와 python을 하다보면 문법이 헷갈릴 때도 있습니다. 예를 들면 이번 ACM-ICPC문제를 한번 python과 ruby로 둘다 풀어봤는데 리스트(루비는 Array)의 길이를 구하는 게 약간씩 다르고, 그래서 오류가 생길 때 한번에 인지하지 못할 때도 있습니다. 그래도 두 언어의 속도나 이런 점을 비교하면 파이썬이 조금 더 유리하다는 생각이 들었습니다. 이번에 이 두 언어를 비교하면서 배우는 것도 도움이 될 것 같습니다.
